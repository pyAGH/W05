```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
```

### Zastosowanie sekwencyjnej metody najmniejszych kwadratów SLSQP (optymalizacja z ograniczeniami) z biblioteki Scipy.optimize
$$ f(x,y)=x^{2}-y^{2}+123$$

#### 1. Definicja funkcji

```python
# Wybór funkcji:
f = lambda x,y: -(x-5)**2+(y-1)**2+123;
#f = lambda x,y: -(x-5)**2+(y-1)**2+123;
```

#### 2. Wykres 3D

```python
figRos = plt.figure(figsize=(12, 7))
axRos = figRos.gca(projection='3d')

X = np.arange(-20, 20, 0.1)
Y = np.arange(-20, 20, 0.1)
X, Y = np.meshgrid(X, Y)
Z = f(X,Y)

surf = axRos.plot_surface(X, Y, Z, cmap=cm.gist_heat_r, linewidth=0, antialiased=False)
#axRos.set_zlim(0, 200)
figRos.colorbar(surf, shrink=0.5, aspect=10)
plt.show()
```

```python
from scipy.optimize import minimize
```

```python
# Sekwencyjna metoda najmniejszych kwadratów (SLSQP)

#Wybór punktu startowego:
x0 = np.array([-5,-5])
#x0 = np.array([5,5])

constraint1 = lambda x: x-3; # Definicja ograniczenia 1: poszukiwnie minimum ma się odbywać tylko na obszarze gdzie x>3
constraint2 = lambda y: y+500; 

con1 = {'type': 'ineq', 'fun': constraint1} # określenie rodzaju ograniczenia (<> lub ==)
con2 = {'type': 'ineq', 'fun': constraint2} # określenie rodzaju ograniczenia (<> lub ==)

cons=[con1,con2]
sol = minimize(f, x0[0],x0[1], method='SLSQP',constraints=cons) 
print(sol)
```
