### Zastosowanie metody złotego podziału do poszukwania minimum funkcji:

```python
import numpy as np 
import matplotlib.pyplot as plt
import math

# Wybór funkcji, dla której chcemy odnaleźć minimum
def f(x):
    #return x**4-5*x**3-2*x**3+5*x-23
    #return x**2
    return -2*x**3+5*x+4

y=f(x)
plt.plot(x,y)
plt.grid()
plt.show()
```

```python
# https://www.programmersought.com/article/33784785839/

def goldenOpt(a,b,Theta_error):
    r=(math.sqrt(5)-1)/2
    a1=b-r*(b-a)
    a2=a+r*(b-a)
    stepNum=0
    y=f(x)
    plt.plot(x,y)
    plt.plot(a,f(a),'g*')
    plt.plot(b,f(b),'g*')
    plt.grid()
    while abs(b-a)>Theta_error:
        stepNum=stepNum+1
        f1=f(a1)
        f2=f(a2)
        if f1>f2:
            a=a1
            f1=f2
            a1=a2
            a2=a+r*(b-a)
        else:
            b=a2
            a2=a1
            f2=f1
            a1=b-r*(b-a)
        x_opt=(a+b)/2
        f_opt=f(x_opt)
        
        plt.plot(x_opt,f_opt,'g*')
        print(x_opt,f_opt)
    plt.show()
    return (x_opt,f_opt,stepNum)

# Określenie zakresu poszukiwań minimum
a=-2
b=2

# Określenie dopuszczalnego błędu
Theta_error=0.3

y=f(x)
plt.plot(x,y)
plt.plot(x_opt,f_opt,'r*')
plt.grid()
plt.show()
x_opt,f_opt,stepNum=goldenOpt(a,b,Theta_error)
x=np.linspace(-2,2,100)


print("Minimum funkcji: ",x_opt)
print("Liczba kroków poszukiwań: ",stepNum)
```
