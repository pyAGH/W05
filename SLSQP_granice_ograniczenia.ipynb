```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import optimize
from matplotlib import cm
```

```python
def f1(x):
    return x[0]**2+x[1]**2+123

def f2(x,y):
    return x**2+y**2+123

x=np.arange(-100,100,1)
y=np.arange(-100,100,1)   
X,Y = np.meshgrid(x,y)

Z = f2(X,Y)
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 500)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');

fig2 = plt.figure()
cset = plt.contourf(X, Y, Z)
```

```python
# Optymalizacja z granic

from scipy.optimize import minimize
x=np.array([2,3])


bnds = ((25, 75), (2, 30)) # Okrelenie granic dla argumentw funkcji optymalizujcej: 25<x[0]<75, 2<x[1]<30


sol = minimize(f1,x, method='SLSQP', bounds=bnds)


# def constraint1(x): # Okrelenie ograniczenia 1 dla funkcji optymalizacyjnej
#     return x[0]-x[1] # x[0]=x[1]

# con1 = {'type': 'eq', 'fun': constraint1} # okrelenie rodzaju ograniczenia (<> lub ==)

# cons=[con1]

#sol = minimize(f1,x, method='SLSQP', bounds=bnds, constraints=cons) # wywoanie funkcji optymalizujcej


print(sol)

```

```python
# Optymalizacja bez granica oraz ogranicze

from scipy.optimize import minimize

x0=np.array([2,3])

sol = minimize(f1,x0, method='SLSQP') # wywoanie funkcji optymalizujcej
print(sol)

```

```python
# Optymalizacja z granic i ograniczeniem

from scipy.optimize import minimize
x=np.array([2,3])


bnds = ((25, 75), (2, 30)) # Okrelenie granic dla argumentw funkcji optymalizujcej: 25<x[0]<75, 2<x[1]<30

def constraint1(x): # Okrelenie ograniczenia 1 dla funkcji optymalizacyjnej
    return x[0]-3*x[1] # x[0]-x[1]   =   0  --> x[0]=3*x[1]

con1 = {'type': 'eq', 'fun': constraint1} # okrelenie rodzaju ograniczenia (<> lub ==)

cons=[con1]

sol = minimize(f1,x, method='SLSQP', bounds=bnds, constraints=cons) # wywoanie funkcji optymalizujcej


print(sol)
```
